```python
%%writefile app.py
import streamlit as st
import pandas as pd
import numpy as np
import pandas_datareader as pdr

apptitle = 'Market Risk - Continous Monitoring'
st.set_page_config(page_title=apptitle, page_icon=":eyeglasses:")
st.image('https://upload.wikimedia.org/wikipedia/commons/thumb/9/98/Capital_One_logo.svg/2560px-Capital_One_logo.svg.png', width=200)
st.title('Macroeconomic Indicators & Market Data Continous Monitoring')


st.markdown("""
 * Use the menu at left to select data and set plot parameters
 * Your plots will appear below


""")


select_graph = st.sidebar.selectbox('What data are you looking for?',
                                    ['None', '3Y/10Y Spreads', '2Y/10Y Spreads', 'Case Shiller Home Price Index', 'Real GDP', 'Unemployment Rate',
                                    'VIX', 'S&P 500', 'CRE', 'CPI'])
dataFilter = st.sidebar.selectbox('Would you like to set a date range for this graph?', [
  'None', 'Yes', 'No'
])

if dataFilter == 'Yes':
  startDate = st.text_input('Enter Start Date', 'YYYY-MM-DD')
  endDate = st.text_input('Enter End Date', 'YYYY-MM-DD')

  if select_graph == 'CPI':
    st.title('CPI Percent Change Over Time from '+ startDate + " to " + endDate)
    chart_data = pdr.DataReader('CPIAUCSL', 'fred')
    chart_data = chart_data.pct_change(periods=12)
    chart_data = chart_data.iloc[12:]
    chart_data['CPIAUCSL'] = chart_data['CPIAUCSL'] *100
    chart_data = chart_data.loc[startDate:endDate]

    st.line_chart(chart_data)



if select_graph == '3Y/10Y Spreads':
  st.title('3Y/10Y Spreads')
  chart_data = pdr.DataReader('T10Y3M', 'fred')

  st.line_chart(chart_data)


if select_graph == '2Y/10Y Spreads':
  chart_data = pdr.DataReader('T10Y2Y', 'fred')

  st.line_chart(chart_data)


if select_graph == 'Case Shiller Home Price Index':
  st.title('Case Shiller Home Price Index')
  chart_data = pdr.DataReader('SPCS20RSA', 'fred')
  chart_data = chart_data.pct_change(periods=12)
  chart_data = chart_data.iloc[12:]
  chart_data['SPCS20RSA'] = chart_data['SPCS20RSA'] *100

  st.line_chart(chart_data)


if select_graph == 'Real GDP':
  st.title('Real GDP (NOTE THIS DATA IS INCORRECT WILL GO BACK AND FIX)')
  chart_data = pdr.DataReader('A191RL1A225NBEA', 'fred')

  st.line_chart(chart_data)

if select_graph == 'Unemployment Rate':
  st.title('Unemployment Rate')
  chart_data = pdr.DataReader('UNRATE', 'fred')

  st.line_chart(chart_data)

if select_graph == 'VIX':
  st.title('VIX - Daily Index Value')
  chart_data = pdr.DataReader('VIXCLS', 'fred')

  st.line_chart(chart_data)

if select_graph == 'S&P 500':
  st.title('S&P 500 - Daily Index Value')
  chart_data = pdr.DataReader('SP500', 'fred')

  st.line_chart(chart_data)

if select_graph == 'CRE':
  st.title('CRE Percent Change Over Time')
  chart_data = pdr.DataReader('BOGZ1FL075035503Q', 'fred')
  chart_data = chart_data.pct_change(periods=4)
  chart_data = chart_data.iloc[4:]
  chart_data['BOGZ1FL075035503Q'] = chart_data['BOGZ1FL075035503Q'] *100

  st.line_chart(chart_data)

if select_graph == 'CPI':
  st.title('CPI Percent Change Over Time')
  chart_data = pdr.DataReader('CPIAUCSL', 'fred')
  chart_data = chart_data.pct_change(periods=12)
  chart_data = chart_data.iloc[12:]
  chart_data['CPIAUCSL'] = chart_data['CPIAUCSL'] *100

  st.line_chart(chart_data)
```
